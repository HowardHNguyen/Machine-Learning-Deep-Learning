# Stratified Spliting

```python
numZeros = []
numZerosStrat = []
expectedZeros = 0.2*list(y).count(0)

# removing random states
stratSplit = StratifiedShuffleSplit(n_splits=1, test_size=0.2)

for i in range(100):
  _,_,_,y_test_temp = train_test_split(X, y, test_size=0.2)
  numZeros.append(list(y_test_temp).count(0))
  for train_index, test_index in stratSplit.split(X, y):
    _,y_test_temp_strat = y[train_index], y[test_index]

  numZerosStrat.append(list(y_test_temp_strat).count(0))
```

```python
import matplotlib.pyplot as plt

plt.bar(expectedZeros, width=0.8, color="green", height=110, label="expected")
plt.hist([numZeros, numZerosStrat], label=["non-strat", "strat"], bins = np.arange(5,15,1), align="left")
plt.legend()
plt.xlabel("Number of zeros")
plt.title("Test set label distribution strat. vs. non-strat. split")
plt.show()
```

# Data Preparation and Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
iris = load_iris()
dfData = pd.DataFrame(np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
print(iris.keys())
print(dfData.head())
```

```python
dfData.isnull().sum()
```

```python
# what type of data do we have? do we have categorical or ordical data in addition to numerical?
dfData.info()
```

```python
# create a train and test sets
from sklearn.model_selection import StratifiedShuffleSplit

stratSplit = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=0)

for train_index, test_index in stratSplit.split(dfData[iris['feature_names']], dfData['target']):
  X_train, X_test = dfData[iris['feature_names']].iloc[train_index], dfData[iris['feature_names']].iloc[test_index]
  y_train, y_test = dfData['target'].iloc[train_index], dfData['target'].iloc[test_index]
```

```python
X_train.isnull().sum()
```

```python
y_train.isnull().sum()
```

```python
# relation plot between all our columns
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
pd.plotting.scatter_matrix(X_train[iris["feature_names"]], ax=ax,c=y_train.values,alpha=0.5)
plt.show()
```

```python
# color coding scatte plot of 2 of our columns
for i in [int(x) for x in y_train.unique().tolist()]:
  plt.scatter(x=X_train.loc[y_train==i]["petal width (cm)"],
              y=X_train.loc[y_train==i]["petal length (cm)"], label=iris["target_names"][i])
plt.xlabel("petal width (cm)")
plt.ylabel("petal length (cm)")
plt.title("petal width vs. petal length")
plt.legend()
plt.show()
```

```python
# correlation, not very useful for non-numercial or non-ordinal categorical targets
pd.concat([X_train, y_train], axis=1).corr().style.background_gradient(cmap="summer")
```

```python
dfTrain = pd.concat([X_train, y_train], axis=1)
labels = dfTrain["target"].unique().tolist()
for label in labels:
  dfTrain[str(iris["target_names"][int(label)])] = dfTrain["target"].apply(lambda x: 1 if x ==label else 0)

# show both ways
# dfTrain.corr().style.background_gradient(cmap="summer")

# show single one
dfTrain.corr()[["setosa","versicolor","virginica"]].style.background_gradient(cmap="summer")
```

# Feature Engineering

```python
dfTrain["petal_ratio"] = dfTrain["petal width (cm)"]/dfTrain["petal length (cm)"]
dfTrain["sepal_ratio"] = dfTrain["sepal width (cm)"]/dfTrain["sepal length (cm)"]
dfTrain.corr()[["setosa","versicolor","virginica"]].style.background_gradient(cmap="summer", low=2)
```

# In this matrix, the values range from -1 to 1, where:

 - 1 indicates a perfect positive correlation,
 - -1 indicates a perfect negative correlation,
 - 0 indicates no correlation.

Here's a brief interpretation of each key component:

1 - Sepal Length (cm):

  - Negatively correlated with the "setosa" class (-0.710066) but positively correlated with the "virginica" class (0.672694). This suggests that setosa typically has shorter sepals, while virginica has longer ones.
  - Weak correlation with "versicolor" (0.037372), indicating less distinction in sepal length for this species.

2 - Sepal Width (cm):

  - Positively correlated with "setosa" (0.577133), indicating that setosa has wider sepals.
  - Negatively correlated with "versicolor" (-0.475208), suggesting that versicolor has narrower sepals.

3 - Petal Length (cm):

  - Strong negative correlation with "setosa" (-0.916907), indicating very short petals in setosa.
  - Strong positive correlation with "virginica" (0.724988), suggesting longer petals in virginica.

4 - Petal Width (cm):

  - Similar to petal length, there's a strong negative correlation with "setosa" (-0.888894) and a strong positive with "virginica" (0.770837).

5 - Target:

  - Strong negative correlation with "setosa" (-0.866025) and positive with "virginica" (0.866025). This might indicate a measure or classification that distinctly identifies setosa and virginica.

6 - Inter-species Correlations:

  - Each species (setosa, versicolor, virginica) is perfectly correlated with itself (1.0) and negatively with the other two (-0.5), which is expected in a one-hot encoding scenario where each species is a distinct class.

7 - Petal Ratio:

  - Negative correlation with "setosa" (-0.825941) and positive with "virginica" (0.604300), indicating some metric where setosa differs significantly from virginica.

8 - Sepal Ratio:

  - High positive correlation with "setosa" (0.900719) and negative with both "versicolor" (-0.398160) and "virginica" (-0.502559), suggesting a distinct trait of setosa in terms of this ratio.

Overall, this matrix illustrates how different measurements and ratios can distinguish among the three Iris species, especially setosa, which appears quite distinct in many metrics.
